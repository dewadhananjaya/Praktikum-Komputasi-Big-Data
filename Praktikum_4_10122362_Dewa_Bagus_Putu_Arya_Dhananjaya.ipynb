menampilkan stastic total semen per kg pada bangunan

```python
sns.set({'figure.figsize':(10,6)}, rc={'axes.facecolor':'white'} )
plt.hist(x=data['cement'], color='skyblue')
plt.title('Total Semen (KG) Pada Pembangunan',size=18)
plt.xlabel('Semen (KG)', size=14)
plt.show()
```

menampilkan stastic kekuatan bangunan dan total nya dengan format MPa

```python
sns.set({'figure.figsize':(10,6)}, rc={'axes.facecolor':'white'} )
plt.hist(x=data['strength'], color='orange')
plt.title('Total Kekuatan Bangunan (MPa)',size=18)
plt.xlabel('Kekuatan Bangunan (MPa)', size=14)
plt.show()
```

2. Jelaskan kesimpulan yang bisa di ambil dari analysis yg anda lakukan

> # Buat uji hipotesis untuk menentukan null atau alternate hypothesis



```python
sns.set({'figure.figsize':(10,6)}, rc={'axes.facecolor':'white'} )
plt.scatter(x=data['cement'], y=data['strength'], color='purple')
plt.title('Hubungan antara Total Cement(KG) dan Kekuatan Bangunan (MPa)',size=18)
plt.xlabel('Semen (KG)', size=14)
plt.ylabel('Kekuatan (MPa)', size=14)
plt.show()
```

Secara umum, null hypothesis akan menyatakan bahwa kedua populasi yang diuji tidak memiliki perbedaan yang signifikan secara statistik.
alternate hypothesis akan menyatakan bahwa ada perbedaan.
Two sampled T-test :-Uji Independent Samples t Test atau 2-sample t-test membandingkan rata-rata dua kelompok independen untuk menentukan apakah ada bukti statistik bahwa rata-rata populasi terkait berbeda secara signifikan.
